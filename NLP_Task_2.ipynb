```python
import pandas as pd
```

```python
df = pd.read_fwf('https://raw.githubusercontent.com/karanpk02/Task_8/main/sample.txt')
df
```

```python
# df.to_csv('sample.txt', index=False)
```

```python
df = df.sample(frac=0.001)
```

```python
df.shape
```

```python
import nltk
from nltk.metrics.distance import jaccard_distance
from nltk.util import ngrams
```

```python
nltk.download('words')
from nltk.corpus import words
correct_words = words.words()
```

```python
# df.values
df = df['2'].tolist()
```

```python
df
```

```python
# list of incorrect spellings
# that need to be corrected 

df1=[]
incorrect_words=['happpy', 'azmaing', 'intelliengt']  
# loop for finding correct spellings
# based on jaccard distance
# and printing the correct word
for word in incorrect_words:
    temp = [(jaccard_distance(set(ngrams(word, 2)),
                              set(ngrams(w, 2))),w)
            for w in correct_words if w[0]==word[0]]
    df1.append(sorted(temp, key = lambda val:val[0])[0][1])        
df1
```

```python
df1=[]
for word in df:
    temp = [(jaccard_distance(set(ngrams(word, 2)),
                              set(ngrams(w, 2))),w)
            for w in correct_words if w[0]==word[0]]
    df1.append(sorted(temp, key = lambda val:val[0])[0][1])        
df1
```

```python
df = pd.DataFrame({'Actual': df, 'Corrected': df1})
df
```
